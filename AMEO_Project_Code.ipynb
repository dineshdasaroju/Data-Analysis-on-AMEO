### Outlier Detection

```python
numerical_Data = Data.select_dtypes("int","float")
```

```python
def outlier_controler(Dataset):
    Q1 , Q3 = np.percentile(Dataset , [25,75])
    IQR = Q3 - Q1
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    return lower_bound,upper_bound
```

```python
for i in numerical_Data:
    lower_bound,upper_bound = outlier_controler(numerical_Data[i])
    
    num2 = numerical_Data.loc[(numerical_Data[i] >= lower_bound) & (numerical_Data[i] <= upper_bound)]
```

```python
num2
```

### Outliers of Numerical Data

```python
for i in numerical_Data:
    lower_bound,upper_bound = outlier_controler(numerical_Data[i])
    
    outlier_num2 = numerical_Data.loc[(numerical_Data[i] <= lower_bound) | (numerical_Data[i] >= upper_bound)]
```

```python
outlier_num2
```

```python
for i in numerical_Data:
    lower_bound,upper_bound = outlier_controler(numerical_Data[i])
    
    Data2 = Data.loc[(Data[i] >= lower_bound) & (Data[i] <= upper_bound)]
```

```python
Data2
```

### Outliers of Whole Data

```python
for i in numerical_Data:
    lower_bound,upper_bound = outlier_controler(numerical_Data[i])
    
    outlier_Data2 = Data.loc[(Data[i] <= lower_bound) | (Data[i] >= upper_bound)]
```
